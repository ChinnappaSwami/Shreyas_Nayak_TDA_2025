```python
from pathlib import Path
import nbformat

# Load the uploaded notebook
notebook_path = Path("Shreyas_Nayak_240911006_week2.ipynb")
with notebook_path.open() as f:
    notebook = nbformat.read(f, as_version=4)

# Clean the notebook
cleaned_cells = []

# Variable renaming map
var_rename_map = {
    "x": "data_x",
    "y": "data_y",
    "model": "regressor_model"
}

def rename_variables(code):
    for old_var, new_var in var_rename_map.items():
        code = code.replace(f"{old_var} ", f"{new_var} ")
        code = code.replace(f"{old_var}=", f"{new_var}=")
        code = code.replace(f"{old_var}.", f"{new_var}.")
        code = code.replace(f"({old_var})", f"({new_var})")
    return code

for cell in notebook.cells:
    if cell.cell_type == 'code':
        # Remove comments and post-output notes
        lines = cell.source.splitlines()
        new_lines = []
        for line in lines:
            stripped = line.strip()
            if stripped.startswith("#") or stripped == "":
                continue
            if stripped.lower().startswith("output") or "output" in stripped.lower():
                continue
            new_lines.append(line)

        cleaned_code = "\n".join(new_lines)
        renamed_code = rename_variables(cleaned_code)
        cleaned_cells.append(nbformat.v4.new_code_cell(source=renamed_code))

# Replace notebook cells
notebook.cells = cleaned_cells

# Save cleaned notebook
cleaned_path = "Cleaned_Shreyas_Nayak_Week2.ipynb"
with open(cleaned_path, "w") as f:
    nbformat.write(notebook, f)

cleaned_path
```
