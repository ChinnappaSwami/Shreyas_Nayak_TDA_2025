## Question 7
#### Fill missing values in the country column with "Unknown" and in the director column with "Not Available"

````python
df["country"].fillna("Unknown",inplace = True)
df["director"].fillna("Not Available" , inplace = True)
df.isnull().sum()
````

## Question 8
#### Replace "United States" with "USA" in the country column.

````python
df["country"] = df["country"].replace("United States", "USA")
df.head()
````

## Question 9
#### Create a new column content_type that contains "Movie" if duration contains "min" and "TV Show" if it contains "Season".

````python
lis = []

for i in df["duration"]:
    if "min" in i :
        lis +=["Movie"]
    elif "Seasons" in i or "Season" in i :
        lis += ["TV Show"]


df["content_type"] = lis

df.head()
````

## Question 10
#### Add a column years_since_release that shows how many years have passed since the release year (assuming current year is 2025).

````python
lis2 = []

for i in df["release_year"] :
    lis2+= [(2025 - i)]

df["years_since_release"] = lis2

df.head()
````

## Question 11
#### Group the dataset by release_year and count how many shows were released each year.

````python
df[df["content_type"] == "TV Show"].groupby("release_year").size().sort_index(ascending=False)
````
